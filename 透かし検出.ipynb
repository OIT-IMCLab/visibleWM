```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from tensorflow.keras.models import  Model
from urllib import request

import matplotlib.pyplot as plt
import os
import copy
import cv2
import numpy as np

model = load_model('WMWRN28_10.h5')
tx_test = np.load('tx_test.npy')
tx_test = tx_test.astype('float32')/255
preds = model.predict(tx_test) #X

for i in range(3):
    plt.imshow(tx_test[i])
    plt.show()
    plt.clf()
    print(i, preds[i])
    print("################################################")
#csv
np.savetxt('WMimagenet.csv', preds, delimiter=',')
```

```python
def noisy(noise_typ,image):
    if noise_typ == "gauss":
        row,col,ch= image.shape
        mean = 0
        sigma = 0.1
        gauss = np.random.normal(mean,sigma,(row,col,ch))
        gauss = gauss.reshape(row,col,ch)
        noisy = image + gauss
        return noisy
    elif noise_typ == "s&p":
        row,col,ch = image.shape
        s_vs_p = 0.5
        amount = 0.004
        out = np.copy(image)
        # Salt mode
        num_salt = np.ceil(amount * image.size * s_vs_p)
        coords = [np.random.randint(0, i - 1, int(num_salt))
                  for i in image.shape]
        out[coords] = 1

        # Pepper mode
        num_pepper = np.ceil(amount* image.size * (1. - s_vs_p))
        coords = [np.random.randint(0, i - 1, int(num_pepper))
                  for i in image.shape]
        out[coords] = 0
        return out
    elif noise_typ == "poisson":
        vals = len(np.unique(image))
        vals = 2 ** np.ceil(np.log2(vals))
        noisy = np.random.poisson(image * vals) / float(vals)
        return noisy
    elif noise_typ =="speckle":
        row,col,ch = image.shape
        gauss = np.random.randn(row,col,ch)
        gauss = gauss.reshape(row,col,ch)        
        noisy = image + image * gauss
        return noisy
```

```python
#

nx_test = np.load('tx_test.npy')
for i in range(5000):
    #
    nx_test[i]=noisy("gauss",nx_test[i])

nx_test=  nx_test.astype('float32')/255
    
print("")

preds = model.predict(nx_test) #X

for i in range(10):
    plt.imshow(nx_test[i])
    plt.show()
    plt.clf()
    print(i, preds[i])
    print("################################################")

np.savetxt('noiseWM.csv', preds, delimiter=',')
```
